### Carregando a base de dados:


```python
import pandas as pd

estabelecimentos = pd.read_csv("20250401_atividade_economica.csv", sep=";")
estabelecimentos
```

### Filtrando bares e restaurantes

```python
bares_restaurantes = estabelecimentos[estabelecimentos['DESCRICAO_CNAE_PRINCIPAL'].str.contains('restaurante', case=False) | estabelecimentos['DESCRICAO_CNAE_PRINCIPAL'].str.contains('bar',case=False)]

bares_restaurantes['DESC_LOGRADOURO'] = bares_restaurantes['DESC_LOGRADOURO'].str.replace(r'\bAVE\b', 'AVENIDA', regex=True)
bares_restaurantes['DESC_LOGRADOURO'] = bares_restaurantes['DESC_LOGRADOURO'].str.replace(r'\bROD\b', 'RODOVIA', regex=True)
bares_restaurantes['DESC_LOGRADOURO'] = bares_restaurantes['DESC_LOGRADOURO'].str.replace(r'\bALA\b', 'ALAMEDA', regex=True)
bares_restaurantes['DESC_LOGRADOURO'] = bares_restaurantes['DESC_LOGRADOURO'].str.replace(r'\bPCA\b', 'PRACA', regex=True)
bares_restaurantes['DESC_LOGRADOURO'] = bares_restaurantes['DESC_LOGRADOURO'].str.replace(r'\bBEC\b', 'BECO', regex=True)
bares_restaurantes['DESC_LOGRADOURO'] = bares_restaurantes['DESC_LOGRADOURO'].str.replace(r'\bTRV\b', 'TREVO', regex=True)

bares_restaurantes['DESC_LOGRADOURO'].unique()
bares_restaurantes
```

### Adicionando coordenadas geográficas

```python
from OSMPythonTools.nominatim import Nominatim


nominatim = Nominatim()

def get_lat_lon(address):
    #EX: "221B Baker Street, London"

    result = nominatim.query(address)

    # Depending on the result, OSMPythonTools may return a single object or a list of objects.
    if isinstance(result, list):
        # If multiple results are returned, pick the first one (or choose based on your criteria)
        osm_object = result[0]
    else:
        osm_object = result
    return osm_object.toJSON()[0]['lat'], osm_object.toJSON()[0]['lon'] 

```

### Atualizando a tabela com coordenadas encontradas automaticamente


```python



nao_encontrados = []
for index, row in bares_restaurantes.iterrows():
    endereco = str(int(row['NUMERO_IMOVEL']))+ ", "+ row['NOME_LOGRADOURO'] + ", Belo Horizonte, MG, Brasil"
    try:
        lat, lon = get_lat_lon(endereco)
        bares_restaurantes.at[index, 'LATITUDE'] = lat
        bares_restaurantes.at[index, 'LONGITUDE'] = lon
    except Exception:
        try:
            endereco = str(int(row['NUMERO_IMOVEL']))+ ", "+row['DESC_LOGRADOURO']+" "+ row['NOME_LOGRADOURO'] + ", Belo Horizonte, MG, Brasil"
            lat, lon = get_lat_lon(endereco)
            bares_restaurantes.at[index, 'LATITUDE'] = lat
            bares_restaurantes.at[index, 'LONGITUDE'] = lon
        except Exception:
            try:
                endereco = str(int(row['NUMERO_IMOVEL']))+ ", "+row['DESC_LOGRADOURO']+" "+ row['NOME_LOGRADOURO'] +", "+row['NOME_BAIRRO']+ ", Belo Horizone, MG, Brasil"
                lat, lon = get_lat_lon(endereco)
                bares_restaurantes.at[index, 'LATITUDE'] = lat
                bares_restaurantes.at[index, 'LONGITUDE'] = lon
            except Exception:
                nao_encontrados.append(endereco)
                print(f"Erro: {endereco}:")

```

### Escrevendo os endereços com coordenadas faltantes em um arquivo
Esses endereços foram colocados na tabela manualmente

```python
f = open("faltantes.txt","w")
for endereco in nao_encontrados:
    f.write(endereco+'\n')
f.close()
```

### Salvando tabela incompleta em um CSV para tratamento

```python
bares_restaurantes.to_csv('bares_restaurantes.csv', sep=';', index=False)
bares_restaurantes
```
