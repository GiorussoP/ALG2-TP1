## Tratamento dos dados via atributo GEOMETRIA

#### Carregando base de dados e filtrando bares/restaurantes

```python
import pandas as pd

estabelecimentos = pd.read_csv("../dados/20250401_atividade_economica.csv", sep=";")
df = estabelecimentos[estabelecimentos['DESCRICAO_CNAE_PRINCIPAL'].str.contains('restaurante', case=False) | estabelecimentos['DESCRICAO_CNAE_PRINCIPAL'].str.contains('bar',case=False)]
```

#### Definição da função de conversão de coordenadas

```python
from pyproj import Transformer

def utm_to_latlon(easting, northing, zone_number=23, northern_hemisphere=False):
    if northern_hemisphere:
        utm_epsg = 32600 + zone_number
    else:
        utm_epsg = 32700 + zone_number

    transformer = Transformer.from_crs(f"EPSG:{utm_epsg}", "EPSG:4326", always_xy=True)

    lon, lat = transformer.transform(easting, northing)
    return lat, lon
```

#### Conversão das coordenadas

```python
import geopandas as gpd
from shapely import wkt

df['GEOMETRIA'] = df['GEOMETRIA'].apply(wkt.loads)

for index, row in df.iterrows():
    row['LATITTUDE'], row ['LONGITUDE'] = utm_to_latlon(row['GEOMETRIA'].x,row['GEOMETRIA'].y,zone_number=23,northern_hemisphere=False)
```

#### Salvando em um arquivo CSV

```python
df.to_csv('../dados/bares_restaurantes.csv', sep=';', index=False)
df
```
